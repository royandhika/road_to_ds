## Correlation  
Pearson Correlation 

```python
pearson_coef, p_value = stats.pearsonr(df['citympg'], df['price'])
print(f'Correlation citympg dengan price')
print(f'Nilai Pearson coefficient adalah {pearson_coef} \ndan p-valuenya adalah {p_value}')
```

```python
pearson_coef, p_value = stats.pearsonr(df['horsepower'], df['price'])
print(f'Correlation horsepower dengan price')
print(f'Nilai Pearson coefficient adalah {pearson_coef} \ndan p-valuenya adalah {p_value}')
```
