```python
def quant_segmen(store):
    if store["Quantity"] <= 3 :
        return "Low"
    elif store["Quantity"] > 3 and store["Quantity"] < 7 :
        return "Medium"
    else :
        return "High"
```

```python
store["Sales Rate"] = store.apply(quant_segmen, axis = 1)
```

### Insert Kolom

```python
sampl = pd.DataFrame({
                    "Nama" : ["Joni", "Jono", "Ucup"],
                    "Umur" : [24, 25, 19]
                    })

sampl.insert(2, "Tinggi",  [180, 175, 177])

sampl
```

### Dengan Fungsi

```python
def stars(row):
    if row.Nama == 'Joni':
        return 3
    elif row.Tinggi > 175:
        return 3
    elif row.Tinggi > 160:
        return 2
    else:
        return 1

tinggi_ratings = sampl.apply(stars, axis='columns')
sampl["Rating"] = tinggi_ratings
sampl
```

```python
df1 = pd.DataFrame({'A':['A0', 'A1', 'A2'], 
                    'B':['B0', 'B1', 'B2'],
                    'age':[24, 27, 20]})
df2 = pd.DataFrame({'A2':['A0', 'A1'],
                    'C':['C0', 'C1'],
                    'D':['D0', 'D1'],
                    'age':[25, 24]})
display(df1)
df2
```

## Join  
Bentuk paling sederhana, menggunakan index dan sebatas kolom saja

![image info](https://miro.medium.com/max/1225/1*-I_1qa5TIiB5eNYxnodfAA.png)

```python
#DataFrame.join(self, other, on=None, how='left', lsuffix='', rsuffix='', sort=False)
df1.join(df2, how = 'left', lsuffix='_x', rsuffix='_y')
# inner, outer, left, right, cross
```

```python
df1.set_index('age').join(df2.set_index('age'), how = 'left', on = 'age')
```

## Merge  
Lebih advance dari join namun masih sebatas kolom

```python
# DataFrame.merge(self, right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None)
df1.merge(df2, how = "outer", on = 'age')
# inner, outer, left, right
```

```python
df1.merge(df2, how = "inner", left_on = "A", right_on = 'A2')
```

## Concat

```python
#pandas.concat(objs, axis=0, join='outer', join_axes=None, ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=None, copy=True)
pd.concat([df1.set_index('age'), df2.set_index('age')], axis = 1, join = 'outer')
```

```python
pd.concat([df1, df2], axis = 1, join = 'outer')
```

```python
pd.concat([df1, df2], axis = 0, join = 'outer')
```

## Binning / Divide

```python
store['Sales'] = store['Sales'].astype('float')
```

```python
salesbin = np.linspace(store['Sales'].min(), store['Sales'].max(), 4)
salesbinname = ['Low', 'Medium', 'High']
store['Sales_binned'] = pd.cut(store['Sales'], salesbin, labels= salesbinname, include_lowest=True)
```

```python
store[store['Sales_binned'] == 'Low']
```
