```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

## Bar

```python
# create a dataframe
data_1 = {'Country':['USA','SGP','JPN','IDN'], 'GDP':[40000,35000,50000,25000] }

df_1 = pd.DataFrame(data_1,columns=['Country','GDP'])
df_1
```

```python
# dengan pandas
df_1.plot(x = 'Country', y = 'GDP', kind = 'bar')
```

```python
# dengan pandas
df_1.plot(x = 'Country', y = 'GDP', kind = 'barh')
```

```python
# dengan matplotlib
plt.figure(figsize=(15,5))
plt.bar(df_1['Country'], df_1['GDP'], width = 0.8)
plt.xlabel('Country')
plt.ylabel('GDP')
plt.title('GDP per Country')
plt.show()
```

```python
# dengan seaborn
plt.figure(figsize=(12,8))
sns.barplot(data = df_1, x = 'Country', y = 'GDP')
plt.title('Country')
plt.show()
```

## Line

```python
# create a dataframe
data_2 = {'Year':[1930,1920,1940,1950], 'GMV':[40000,35000,50000,25000] }

df_2 = pd.DataFrame(data_2,columns=['Year','GMV'])
df_2
```

```python
df_2_sort = df_2.sort_values(by='Year')
df_2_sort
```

```python
# pandas
df_2_sort.plot(x = 'Year', y = 'GMV', kind = 'line')
```

```python
# matplotlib
plt.figure(figsize = (6,6))
plt.plot(df_2_sort['Year'], df_2_sort['GMV'], marker = 'o')
plt.ylim(0)
plt.grid()
plt.xlabel('Year')
plt.ylabel('GMV')
plt.title('GMV every Year', fontsize = 18)
plt.show()
```

```python
# seaborn
plt.figure(figsize = (6,6))
sns.lineplot(data = df_2_sort, x = 'Year', y = 'GMV')
plt.grid()
plt.title('GMV every Year', fontsize = 18)
plt.ylim(0, 51000)
plt.show()
```

## Scatter

```python
## create a dataframe
data_3 = {'Unemployment_Rate': [6.1,5.8,5.7,5.7,5.8,5.6,5.5,5.3,5.2,5.2],
        'Stock_Index_Price': [1500,1520,1525,1523,1515,1540,1545,1560,1555,1565]
       }
  
df_3 = pd.DataFrame(data_3,columns=['Unemployment_Rate','Stock_Index_Price'])
df_3
```

```python
# pandas
df_3.plot(x = 'Unemployment_Rate', y = 'Stock_Index_Price', kind = 'scatter')
```

```python
# matplotlib
plt.figure(figsize = (7,7))
plt.scatter(df_3['Unemployment_Rate'], df_3['Stock_Index_Price'])
plt.grid()
```

```python
# seaborn
sns.relplot(data = df_3, x = 'Unemployment_Rate', y ='Stock_Index_Price')
plt.grid()
plt.title('GMV every Year', fontsize = 15)
plt.show()
```
